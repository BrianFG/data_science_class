```python
bedrooms_6 = house_df.query('bedrooms == 6')['bathrooms']
bedrooms_7 = house_df.query('bedrooms == 7')['bathrooms']
bedrooms_5 = house_df.query('bedrooms == 5')['bathrooms']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scale_columns = ['yr_built', 'yr_renovated', 'zipcode', 'lat', 'long']
house_df[scale_columns] = scaler.fit_transform(house_df[scale_columns])
```

```python
house_df
```

```python
from statsmodels.formula.api import ols
```

```python
formula = f"""price ~ 
id +
bedrooms +
bathrooms +
sqft_living +
sqft_lot +
floors +
waterfront +
view +
condition +
grade +
sqft_above +
sqft_basement +
yr_built +
yr_renovated +
zipcode +
lat +
long +
sqft_living15 +
sqft_lot15"""
results = ols(formula, data=house_df).fit()
```

```python
results.summary()
```

```python
house_df.corr()['price'].sort_values()
```

```python
results.rsquared
```

```python
results.params.sort_values()
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.feature_selection import RFE, f_regression

from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
```

```python
for column in house_df.columns:
    results = ols(f"price ~ {column}", data=house_df).fit()
    print(column, results.rsquared)
    print(results.f_pvalue)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x = house_df.drop(['price'], axis=1)
```

```python
y = house_df['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=40)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
